```python
neet_try_survey.drop(axis=0, index=[2, 5, 8, 11, 14, 17, 20, 23, 26, 29, 32], inplace=True)
```

```python
neet_try_survey
```

```python
grouped_data = neet_try_survey.groupby(['항목', '여부'])['명'].sum().reset_index()

import altair as alt

# Create a bar chart
chart = alt.Chart(grouped_data).mark_bar().encode(
    x=alt.X('여부', title='여부'),  # Map '여부' to the x-axis
    y=alt.Y('명', title='총합'),   # Map '명' to the y-axis
    color='항목',                      # Color the bars by '항목'
    column='항목'                      # Create separate columns for each '항목'
).properties(
    title='항목에 따른 여부별 값' # Set the chart title
)

chart
```
